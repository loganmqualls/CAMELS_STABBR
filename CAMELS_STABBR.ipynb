```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geocoder
from tqdm.notebook import tqdm
```

```python
#path to data source
path = '/home/NearingLab/data/camels_attributes_v2.0/gauge_id_lat_lon_name.txt'
colnames = ['ID', 'lat','lon','location']
df = pd.read_csv(path, sep = ';', names = colnames)
```

```python
#formatting
df = df.set_index('ID')
df = df.drop('location',axis=1)
```

```python
#concatenating lat-lon, just in case... thanks Jonathan
#df["lat-lon"] = ['{:.4f}, {:.4f}'.format(df.loc[i,'lat'], df.loc[i,'lon']) for i in df.index.values]
```

```python
#Querying...Thanks Grey
states = {}
for index, row in tqdm(df.iterrows()):
    geostring = f"{str(row['lat'])}, {str(row['lon'])}"
    counter = 0
    while counter < 10:
        try:
            counter += 1
            query = geocoder.uscensus(geostring, method='reverse')
            states[row.name] = query.json['state']
            break
        except:
            states[row.name] = 'missing'
            continue
```
